```python
#Highest sold Product in top 10 order
df.sort_values(by = "Quantity", ascending=False).head(10)
```

```python
#City that has done best with sale
df2=df.groupby('City')[['Quantity','Total','cogs',
                       'Rating']].mean(numeric_only=True).sort_values(by="Total",ascending=False)
df2 
```

```python
df3 = df2.transpose()
df3
```

```python
df3.plot(kind = 'line',title='Sales of Supermarket',ylabel='Scores')
```

```python
df3.boxplot()
```
